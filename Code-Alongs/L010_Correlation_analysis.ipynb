## Correlation analysis  
Correlation analysis is a statistical technique used to examine the strength and relationship between two or more variables.  
It involves analysing the degree to which changes in one variable are associated with changes in another variable.

### Pandas .corr() method  
Calculates the correlation coeffients between all pairs of features in a dataframe, and returns a new dataframe with where the number of rows and columns both match the number of columns in the input df.

```python
correlation_matrix = titanic.corr()
correlation_matrix
```

```python
# Setting correlation threshold. Anything above this is considered as highly correlated
threshold = 0.85
```

```python
import numpy as np
upper = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k = 1).astype(bool))
upper.style.map(lambda x: "color: red" if x > threshold else "")
```

```python
columns_to_drop = []

for i in range(len(correlation_matrix.columns)):
    column_name = correlation_matrix.columns[i]
    # print(column_name, end=": ")
    
    for j in range(i):
        # print(correlation_matrix.iloc[i, j], end = ", ")
        if correlation_matrix.iloc[i, j] > threshold:
            # row_name = correlation_matrix.columns[j]
            columns_to_drop.append(column_name)
            break
    # print()
print(columns_to_drop)
```

```python
titanic.drop(columns=columns_to_drop)
```

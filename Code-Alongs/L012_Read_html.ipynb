```python
import pandas as pd
import numpy as np
```

```python
boxoffice = pd.read_html("https://en.wikipedia.org/wiki/List_of_Pixar_films", match="Box office gross", header=1)[0]
boxoffice
```

```python
# boxoffice.iloc[22:25, 3] = np.nan
# boxoffice.iloc[23,2] = np.nan
boxoffice.tail()
```

```python
boxoffice.rename(columns={"U.S. and Canada": "US_Canada", "Other territories": "Other"}, inplace=True)
boxoffice.drop(columns=boxoffice.columns[-1], inplace=True)
```

```python
boxoffice.head()
copy = boxoffice.copy()
copy["US_Canada"] = copy["US_Canada"].str.lstrip('$').str.replace(',', '')
copy["Other"] = copy["Other"].str.lstrip('$').str.replace(',', '')
copy["Worldwide"] = copy["Worldwide"].str.lstrip('$').str.replace(',', '')
copy["US_Canada"] = pd.to_numeric(copy["US_Canada"])
copy["Other"] = pd.to_numeric(copy["Other"])
copy["Worldwide"] = pd.to_numeric(copy["Worldwide"])
copy.head()
```
